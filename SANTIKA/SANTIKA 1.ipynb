# Function List

```python
import pandas as pd
import numpy as np
import csv
from matplotlib import pyplot as plt
from datetime import datetime , timedelta
from dateutil.parser import parse

def load_csv(file_name):
    result = []
    with open(file_name,"r") as csvfile:
        reader = csv.reader(csvfile)

        for r in reader:
            result.append(r)
    return np.array(result)

def write_csv(file_name, data):
    with open(file_name, "w") as csvfile:
        writer = csv.writer(csvfile,quoting=csv.QUOTE_ALL)
        writer.writerow(data)

# Get Today spesific time
todayTime = datetime.today()
todayTime = todayTime.strftime("%Y-%m-%d")
```

# Creating data upload pengguna selama seminggu

```python
# Get tanggal minggu lalu
minggu_lalu = datetime.today() - timedelta(days=7)
minggu_lalu = minggu_lalu.strftime("%Y-%m-%d")
# print(todayTime)
# print(minggu_lalu)

countMinggu_awkar = []

awkarin = data[data.username == 'awkarin']    
awkarin = awkarin[awkarin.date >= str(minggu_lalu)]

for (i,item) in enumerate(awkarin.date):
    datetime_object = datetime.strptime(awkarin.date.iloc[i], '%Y-%m-%d %H:%M:%S') #change value to date type
    datetime_object = datetime_object.strftime("%Y-%m-%d") #shortening date type
    countMinggu_awkar.append(datetime_object)
#     print(datetime_object)

hari = "|".join(countMinggu_awkar)
hari = hari.split("|")
haris, freqs = np.unique(hari, return_counts=True)
# print(names)
print("dalam satu minggu terakhir uploads :",len(haris))

print()

awkarin_upload_week = np.array([(dey, freq) for dey,freq in sorted(zip(haris,freqs), reverse=False)])[:7]
# print("Top 7 user dicari: ", most_upload_week)
print(awkarin_upload_week[:,:])
```
